```python
import pandas as pd

df = pd.read_csv('C:\lipophilicity descriptors.csv')
df
```

```python
# Data Preparation
#Separate features (X) and target (y)
y = df['MolLogP']
y
```

```python
X = df.drop('MolLogP',axis=1)
X
```

```python
#Split Data into Training and Testing Sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

```python
#Build and Train a Linear Regression Model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#Applying the model to make prediction
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

```python
#Evaluate model performance
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

```python
#Build and Train a Random Forest Model
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

```python
#Applying the model to make a prediction
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

```python
#Evaluate model performance
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

```python
# Train Gradient Boosting model
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(random_state=42)
gb.fit(X_train, y_train)
```

```python
# Applying the model to make a Prediction
y_gb_train_pred = gb.predict(X_train)
y_gb_test_pred = gb.predict(X_test)
```

```python
# Evaluate model performance
from sklearn.metrics import mean_squared_error, r2_score

gb_train_mse = mean_squared_error(y_train, y_gb_train_pred)
gb_train_r2 = r2_score(y_train, y_gb_train_pred)

gb_test_mse = mean_squared_error(y_test, y_gb_test_pred)
gb_test_r2 = r2_score(y_test, y_gb_test_pred)
```

```python
gbr_results = pd.DataFrame(['Gradient Boosting', gb_train_mse, gb_train_r2, gb_test_mse, gb_test_r2]).transpose()
gbr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
gbr_results
```

```python
# Model comparison
df_models = pd.concat([lr_results, rf_results, gbr_results], axis=0).reset_index(drop=True)
print(df_models)
```

```python
#Data visualization of prediction results
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 6))
# Scatter plot for Linear Regression Predictions
plt.scatter(x=y_train, y=y_lr_train_pred, c="#7CAE00", alpha=0.3, label="Linear Regression")
# Fit a line for Linear Regression
z_lr = np.polyfit(y_train, y_lr_train_pred, 1)
p_lr = np.poly1d(z_lr)
plt.plot(y_train, p_lr(y_train), '#F8766D', linestyle='--')

# Scatter plot for Random Forest Predictions
plt.scatter(x=y_train, y=y_rf_train_pred, c="#00BFC4", alpha=0.3, label="Random Forest")
# Fit a line for Random Forest
z_rf = np.polyfit(y_train, y_rf_train_pred, 1)
p_rf = np.poly1d(z_rf)
plt.plot(y_train, p_rf(y_train), '#F8766D', linestyle='--')

# Scatter plot for Gradient Boosting Predictions
plt.scatter(x=y_train, y=y_gb_train_pred, c="#C77CFF", alpha=0.3, label="Gradient Boosting")
# Fit a line for Gradient Boosting
z_gb = np.polyfit(y_train, y_gb_train_pred, 1)
p_gb = np.poly1d(z_gb)
plt.plot(y_train, p_gb(y_train), '#F8766D', linestyle='--')

# Labels for the axes
plt.xlabel('Experimental MLogP')
plt.ylabel('Predicted MLogP')

# Title for the plot
plt.title('Comparison of Model Predictions')

# Add a legend to differentiate between models
plt.legend()

# Save the plot as PNG and PDF files
plt.savefig('ModelComparison_ScatterPlot.png')
plt.savefig('ModelComparison_ScatterPlot.pdf')

# Display the plot
plt.show()
```
